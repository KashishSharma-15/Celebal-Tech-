```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

df= pd.read_csv("D:/celebal tech intership/cancer_data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.histplot(data=df, x='Diagnosis')
plt.xlabel('Diagnosed')
plt.title('Cancer Diagnosis', color='red')
```

```python
sns.heatmap(df)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Box plot of BMI by Diagnosis
plt.figure(figsize=(10, 6))
df.boxplot(column='BMI', by='Diagnosis', grid=False)
plt.title('Box Plot of BMI by Diagnosis')
plt.suptitle('')  # Suppress the automatic title
plt.xlabel('Diagnosis')
plt.ylabel('BMI')
plt.show()
```

```python
sns.catplot(df, x='BMI', y= 'Age', kind='box', palette ='winter')
```

```python
# Bar plot of Gender
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.show()
```

```python
# Pie chart of Diagnosis
diagnosis_counts = df['Diagnosis'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'skyblue'], startangle=140)
plt.title('Cancer Diagnosis Distribution')
plt.show()
```

```python
# Scatter plot of Age vs. BMI
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['BMI'], c=df['Diagnosis'], cmap='coolwarm', edgecolor='k', alpha=0.7)
plt.title('Scatter Plot of Age vs. BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.colorbar(label='Diagnosis')
plt.show()
```

```python
# Suppress specific warnings
warnings.filterwarnings('ignore', category=FutureWarning, module='seaborn')
warnings.filterwarnings('ignore', category=FutureWarning, module='pandas')


# Pair plot
sns.pairplot(df, hue='Diagnosis', diag_kind='kde')
plt.suptitle('Pair Plot of Cancer Data', y=1.02)
plt.show()

```
