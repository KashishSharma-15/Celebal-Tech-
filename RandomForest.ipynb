```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('Data.csv')
print(df.head)
```

```python
df.describe()
```

```python
df.info()
```

```python
y = df['Column25']
X = df.drop(['Column25'], axis = 1)
```

```python
random_seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= random_seed)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Create a Random Forest classifier with specified parameters
rf_classifier = RandomForestClassifier(n_estimators=100)

# Fit the Random Forest model to the training data
rf_classifier.fit(X_train, y_train)

# Use the trained model to make predictions on the test data
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Assuming y_test contains the true labels and y_pred contains the predicted labels
# Replace y_test and y_pred with your actual variables

# Compute accuracy
accuracy_rf = accuracy_score(y_test, y_pred)

print("Accuracy of Random Forest model:", accuracy_rf)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Compute confusion matrix for Random Forest model
conf_matrix_rf = confusion_matrix(y_test, y_pred)

# Plot confusion matrix for Random Forest model
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix for Random Forest Model')
plt.show()

```

```python
# plot the accuracy of the random forest classifier

# Import necessary libraries
import matplotlib.pyplot as plt

# Create a list to store the number of trees and the corresponding accuracy scores
trees = []
accuracy_scores = []

# Iterate over different number of trees
for n_trees in range(10, 201, 10):
    # Create a new Random Forest classifier with the current number of trees
    rf_classifier = RandomForestClassifier(n_estimators=n_trees)

    # Fit the classifier to the training data
    rf_classifier.fit(X_train, y_train)

    # Predict on the test data
    y_pred = rf_classifier.predict(X_test)

    # Calculate the accuracy score
    accuracy = accuracy_score(y_test, y_pred)

    # Append the number of trees and the accuracy score to the lists
    trees.append(n_trees)
    accuracy_scores.append(accuracy)

# Plot the number of trees vs. accuracy score
plt.figure(figsize=(10, 6))
plt.plot(trees, accuracy_scores, marker='o')
plt.xlabel('Number of Columns')
plt.ylabel('Accuracy Score')
plt.title('Random Forest Accuracy vs. Number of Columns')
plt.show()
```

```python
# bar plot  to show which column is more accurate in predicting the accurate thing

# Calculate the importance of each feature
importances = rf_classifier.feature_importances_

# Sort the features by importance in descending order
features = X.columns
sorted_indices = np.argsort(-importances)
sorted_features = [features[i] for i in sorted_indices]

# Create a bar plot to visualize the importance of each feature
plt.figure(figsize=(10, 6))
plt.bar(sorted_features, importances[sorted_indices], color='pink')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance for Random Forest Model')
plt.show()
```
